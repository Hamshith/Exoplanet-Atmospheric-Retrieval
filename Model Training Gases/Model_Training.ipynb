```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ABC_Feature_Matrix_Random.csv')
df
```

```python
x_spec = df.iloc[:, 1:53]
x_spec
```

```python
y = df.iloc[:, 54:]
y
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_spec, y, test_size = 0.2, random_state = 42)
```

Normalisation of each sample (i.e row wise normalisation)

```python
#Manual Normalisation - numpy
x_train = x_train.to_numpy()
train_means = x_train.mean(axis=1, keepdims=True)
train_stds = x_train.std(axis=1, keepdims=True)
x_train_norm = (x_train - train_means) / train_stds
x_train_norm
```

```python
x_train_enh = np.hstack([x_train_norm, train_means, train_stds])
#Added to so that the model can learn info about info lost during row wise normalisation
```

```python
x_test = x_test.to_numpy()
test_means = x_test.mean(axis=1, keepdims=True)
test_stds = x_test.std(axis=1, keepdims=True)
x_test_norm = (x_test - test_means) / test_stds
```

```python
x_test_enh = np.hstack([x_test_norm, test_means, test_stds])
```

```python
y_train_np = y_train.to_numpy()
y_test_np  = y_test.to_numpy()
```

```python
import xgboost as xgb
from sklearn.multioutput import MultiOutputRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
```

```python
xgb_base = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=500,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)
```

```python

svr_base = SVR(
    kernel='rbf', 
    C = 50,
    epsilon = 0.1
)
```

```python
mlp_base = MLPRegressor(
    hidden_layer_sizes=(100, 50),      # 2 hidden layers: 100 and 50 neurons
    activation='relu',                  # ReLU activation
    solver='adam',                      # Adam optimizer
    alpha=0.0001,                       # L2 regularization
    batch_size='auto',                  # Auto batch size
    learning_rate='adaptive',           # Adaptive learning rate
    learning_rate_init=0.001,          # Initial learning rate
    max_iter=200,                       # Maximum epochs
    random_state=42,
    verbose=True,                       # Show training progress
    early_stopping=True,                # Stop if no improvement
    validation_fraction=0.1,            # 10% for validation
    n_iter_no_change=10                 # Stop after 10 epochs with no improvement
)
```

```python
models = {
    "XGBoost" : MultiOutputRegressor(xgb_base),
    "SVR" : MultiOutputRegressor(svr_base),
    "MLP Regressor" : mlp_base
}
```

CV Scores (r2 scoring) for each of the three base models - XGB, SVR, MLP

```python
from sklearn.model_selection import cross_val_score
for name, model in models.items():
    score = cross_val_score(model, x_train_norm, y_train, cv = 5, scoring = 'r2')
    print(f"{name} : {score.mean()}")
```

```python
from sklearn.model_selection import cross_val_score
for name, model in models.items():
    score = cross_val_score(model, x_train_enh, y_train, cv = 5, scoring = 'r2')
    print(f"{name} : {score.mean()}")
```

```python
#MLP fine tuning using optuna for norm
import optuna
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
import numpy as np
import time

print("="*70)
print("OPTUNA HYPERPARAMETER TUNING FOR MLP REGRESSOR")
print("="*70)

# ============================================================================
# DEFINE OBJECTIVE FUNCTION
# ============================================================================

def objective(trial):
    """
    Objective function for Optuna to optimize.
    Returns the cross-validation R² score.
    """
    
    # 1. ARCHITECTURE: Number of layers and neurons
    n_layers = trial.suggest_int('n_layers', 1, 4)  # 1 to 4 hidden layers
    
    hidden_layers = []
    for i in range(n_layers):
        n_units = trial.suggest_int(f'n_units_l{i}', 32, 256, step=32)
        hidden_layers.append(n_units)
    
    # 2. REGULARIZATION
    alpha = trial.suggest_float('alpha', 1e-5, 1e-2, log=True)
    
    # 3. LEARNING RATE
    learning_rate_init = trial.suggest_float('learning_rate_init', 1e-4, 1e-2, log=True)
    
    # 4. BATCH SIZE
    batch_size = trial.suggest_categorical('batch_size', [32, 64, 128, 'auto'])
    
    # 5. ACTIVATION FUNCTION
    activation = trial.suggest_categorical('activation', ['relu', 'tanh'])
    
    # 6. LEARNING RATE SCHEDULE
    learning_rate = trial.suggest_categorical('learning_rate', ['constant', 'adaptive'])
    
    # 7. EARLY STOPPING PATIENCE
    n_iter_no_change = trial.suggest_int('n_iter_no_change', 5, 20)
    
    # 8. VALIDATION FRACTION
    validation_fraction = trial.suggest_float('validation_fraction', 0.1, 0.2)
    
    # Create model with suggested parameters
    mlp = MLPRegressor(
        hidden_layer_sizes=tuple(hidden_layers),
        activation=activation,
        solver='adam',  # Keep Adam (best for most cases)
        alpha=alpha,
        batch_size=batch_size,
        learning_rate=learning_rate,
        learning_rate_init=learning_rate_init,
        max_iter=300,  # Increase from 200 for better convergence
        random_state=42,
        verbose=False,  # Disable verbose in trials
        early_stopping=True,
        validation_fraction=validation_fraction,
        n_iter_no_change=n_iter_no_change,
        tol=1e-4
    )
    
    # Use 3-fold cross-validation for faster tuning
    try:
        scores = cross_val_score(
            mlp, 
            x_train_norm, 
            y_train, 
            cv=3, 
            scoring='r2',
            n_jobs=-1  # Use all cores
        )
        
        # Return mean CV score
        return scores.mean()
    
    except Exception as e:
        # If training fails, return very low score
        print(f"Trial failed: {e}")
        return -999.0

# ============================================================================
# RUN OPTUNA OPTIMIZATION
# ============================================================================

print("\nStarting Optuna optimization...")
print("This will take approximately 20-40 minutes depending on your hardware.\n")

# Create study
study = optuna.create_study(
    direction='maximize',  # Maximize R²
    sampler=optuna.samplers.TPESampler(seed=42),  # Tree-structured Parzen Estimator
    pruner=optuna.pruners.MedianPruner(  # Prune unpromising trials early
        n_startup_trials=5,
        n_warmup_steps=10
    )
)

# Add a callback to show progress
def callback(study, trial):
    if trial.number % 5 == 0:
        print(f"\nTrial {trial.number} completed:")
        print(f"  Current value: {trial.value:.4f}")
        print(f"  Best value so far: {study.best_value:.4f}")

# Optimize
start_time = time.time()

study.optimize(
    objective, 
    n_trials=30,  # Try 30 different parameter combinations
    timeout=None,  # No timeout
    callbacks=[callback],
    show_progress_bar=True,
    n_jobs=1  # Don't parallelize trials (each trial uses parallel CV)
)

elapsed = time.time() - start_time

print("\n" + "="*70)
print("OPTIMIZATION COMPLETED!")
print("="*70)
print(f"Total time: {elapsed/60:.1f} minutes")
print(f"Best CV R²: {study.best_value:.4f}")
print(f"Number of finished trials: {len(study.trials)}")

# ============================================================================
# BEST PARAMETERS
# ============================================================================

print("\n" + "="*70)
print("BEST HYPERPARAMETERS:")
print("="*70)

best_params = study.best_params.copy()

# Extract hidden layer architecture
n_layers = best_params.pop('n_layers')
hidden_layers = tuple([best_params.pop(f'n_units_l{i}') for i in range(n_layers)])

print(f"\nArchitecture:")
print(f"  hidden_layer_sizes: {hidden_layers}")

print(f"\nRegularization & Optimization:")
for param, value in best_params.items():
    print(f"  {param}: {value}")
```

```python
#XGB optimisation (Enhanced)
import optuna

# Validation split inside training set
X_tr, X_val, y_tr, y_val = train_test_split(
    x_train_enh, y_train, test_size=0.2, random_state=42
)

def objective(trial):

    params = {
        "objective": "reg:squarederror",
        "verbosity": 0,
    
        "n_estimators": trial.suggest_int("n_estimators", 300, 1500),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3),
        "max_depth": trial.suggest_int("max_depth", 3, 12),

        "min_child_weight": trial.suggest_int("min_child_weight", 1, 20),
        "gamma": trial.suggest_float("gamma", 0.0, 1.0),

        "subsample": trial.suggest_float("subsample", 0.4, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.4, 1.0),

        "reg_alpha": trial.suggest_float("reg_alpha", 0.0, 10.0),
        "reg_lambda": trial.suggest_float("reg_lambda", 0.0, 10.0),
    }




    model = MultiOutputRegressor(xgb.XGBRegressor(**params))
    model.fit(X_tr, y_tr)

    preds = model.predict(X_val)

    # mean R² across all outputs
    scores = [
        r2_score(y_val, preds)
        for i in range(y_val.shape[1])
    ]
    mean_r2 = np.mean(scores)

    trial.report(mean_r2, step=0)
    if trial.should_prune():
        raise optuna.exceptions.TrialPruned()

    return mean_r2
```
