```python
import numpy as np
import pandas as pd
```

```python
abc_df = pd.read_csv('ABC_Feature_Matrix_Random.csv')
abc_df
```

```python
y_h2o = abc_df.iloc[:, 54]
```

```python
import matplotlib.pyplot as plt
import scipy.stats as st
plt.hist(y_h2o, bins=50)
plt.title(f"skew={st.skew(y_h2o):.3f} of H2O in ABC dataset ")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(y_h2o, bins=40, alpha=0.6, color='skyblue', edgecolor='k')
plt.axvline(-5.94, color='red', linestyle='--', label= "WASP-39b H20(-5.94)")
plt.title(f"H2O distribution in ABC")
plt.xlabel("log10 abundence")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.show()
```
