```python
import matplotlib.pyplot as plt
import pandas as pd

df_20k = pd.read_csv("ABC_Feature_Matrix_Random.csv")
df_90k = pd.read_csv("complete_dataset.csv")

molecules = ['Temperature','H2O', 'CO2', 'CH4', 'CO', 'NH3']
target_cols = ['planet_temp','log_H2O', 'log_CO2', 'log_CH4', 'log_CO', 'log_NH3']

for mol, col in zip(molecules, target_cols):
    plt.figure(figsize=(5,4))
    plt.hist(df_20k[col], alpha=0.5, label='20k', bins=50, density=True)
    plt.hist(df_90k[col], alpha=0.5, label='90k', bins=50, density=True)
    plt.title(mol)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
